# Load and plot OMNI data

This is a simple notebook to show how to load the OMNI data, which include the solar wind parameters as well as some frequently used geomagnetic indices, and to manipulate and plot the data.

First you should import the PySPEDAS and PyTplot modules. 
Then you can use `omni.data()` to load OMNI data. 



```python
import pyspedas
import pytplot
from pytplot import tplot   ## As a shortcut for pytplot.tplot()

pyspedas.omni.data( trange=['2017-03-24', '2017-03-31'], time_clip=False )  ## Download and then load OMNI data for 8 days from March 24, 2017. 
```

Set a time range for which plots are made using `timespan()`, and plot several tplot variables with `tplot()`. 


```python
pytplot.timespan( "2017-03-24 00:00:00", 7, keyword='days')
tplot(['SYM_H','AL_INDEX','BX_GSE', 'BY_GSM', 'BZ_GSM', 'flow_speed', 'proton_density', 'Pressure'])
```

The keywords `xsize` and `ysize` can be used to resize a plot. The default size is approximately (xsize, ysize) = (8, 9). The example right below thus produces a vertically-long plot. 

```python
tplot( ['SYM_H','flow_speed', 'Pressure'], xsize=4, ysize=10 )
```

`pytplot.options()` module changes various attributes for a tplot variable. The example below changes the type of the vertical axis for the tplot variable `Pressure` to logarithmic. The resultant plot shows the dynamic pressure (bottom panel) with a log scale. 

```python
pytplot.options('Pressure', 'ylog', True )
tplot( ['SYM_H','flow_speed', 'Pressure'], xsize=12, ysize=4 )  ## This pair of (xsize, ysize) would lead to a horizontally-long plot. 
```

The following example changes the title of the tplot variable `Pressure`. 

`ytitle` attribute holds a string that is to be shown by the vertical axis of a tplot variable. '\n' (backslash + n) is replaced with a newline. 

```python
pytplot.options('Pressure', 'ytitle', 'Dynamic\npressure' )
tplot( ['SYM_H','flow_speed', 'Pressure'] )
```

`yrange` attribute should be set to be a two-element array determining the range of a vertical axis. 

`timebar()` can draw a vertical line at a designated date&time. Setting `delete` keyword to be True, the command erases the vertical lines previously drawn at the given date&times.  

```python
pytplot.options( 'AL_INDEX', 'yrange', [-1000,0])
t_lines = pyspedas.time_double('2017-03-26 20:00:00')
pytplot.timebar( t_lines )
tplot( ['SYM_H','AL_INDEX', 'Pressure'] )
```

```python
pytplot.timebar( t_lines, delete=True )
tplot( ['SYM_H','AL_INDEX', 'Pressure'] )
```
