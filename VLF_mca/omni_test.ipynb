```python

import pyspedas
import pytplot
from pytplot.MPLPlotter.tplot import tplot
trange = ['1989-03-10', '1989-4-1']

#mca(trange = trange)
#orb(trange = trange)
omni_vars = pyspedas.omni.data(trange= trange, level='hro', datatype='5min')

pytplot.store_data("IMF", data=['BX_GSE', 'BY_GSM', 'BZ_GSM'])
omni_data_names = ['SYM_H', 'IMF', 'flow_speed', 'proton_density', 'Pressure', 'E']
pytplot.options(omni_data_names, 'panel_size', 0.5)
pytplot.options('IMF', 'legend_names', ['IMF x', "IMF y", "IMF z"])
tplot(['SYM_H', 'flow_speed', 'proton_density', 'Pressure', 'E'], xsize = 16)
```
